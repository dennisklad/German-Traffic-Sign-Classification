```python
import os
import cv2
import matplotlib.pyplot as plt

def get_image_resolution(image_path):
    image = cv2.imread(image_path)
    height, width, _ = image.shape
    return width, height

def show_image_resolution(dataset_dir):
    class_resolutions = {}  # Dictionary to store class resolutions

    # Assume the dataset directory contains subdirectories for each class
    class_dirs = [class_name for class_name in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, class_name))]

    for class_name in class_dirs:
        class_path = os.path.join(dataset_dir, class_name)
        image_files = [img for img in os.listdir(class_path) if img.lower().endswith('.ppm')]

        # Get the resolution of each image in the class
        resolutions = [get_image_resolution(os.path.join(class_path, img)) for img in image_files]

        # Calculate the average resolution for the class
        avg_width = sum([res[0] for res in resolutions]) // len(resolutions)
        avg_height = sum([res[1] for res in resolutions]) // len(resolutions)

        class_resolutions[class_name] = (avg_width, avg_height)

    # Sort the classes based on their average height resolution
    sorted_classes = sorted(class_resolutions.items(), key=lambda x: x[1][1])

    # Plot the image resolutions using subplots
    num_classes = len(sorted_classes)
    num_cols = 5  # Number of columns in the subplot grid
    num_rows = (num_classes + num_cols - 1) // num_cols  # Number of rows in the subplot grid

    plt.figure(figsize=(50, 50))
    for i, (class_name, resolution) in enumerate(sorted_classes, 1):
        plt.subplot(num_rows, num_cols, i)
        plt.bar(class_name, resolution[1])
        plt.xticks([])
        plt.xlabel('Class', fontsize=15)
        plt.ylabel('Average Image Resolution (Height)', fontsize=15)

    plt.suptitle('Average Image Resolution of Each Class', fontsize=20)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Add space for the title
    plt.show()

# Example usage:
PATH = "/srv/store/docker-users/pjstuds/obada.kiddeh" # root directory
dataset_directory = PATH + '/FinalDataset/Final_Training/' # path to the dataset
show_image_resolution(dataset_directory)
```

```python
import os
import cv2
import pandas as pd

def get_image_resolution(image_path):
    image = cv2.imread(image_path)
    height, width, _ = image.shape
    return width, height

def show_image_resolution(dataset_dir):
    data = []  # List to store class resolutions data

    # Assume the dataset directory contains subdirectories for each class
    class_dirs = [class_name for class_name in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, class_name))]

    for class_name in class_dirs:
        class_path = os.path.join(dataset_dir, class_name)
        image_files = [img for img in os.listdir(class_path) if img.lower().endswith('.ppm')]

        # Get the resolution of each image in the class
        resolutions = [get_image_resolution(os.path.join(class_path, img)) for img in image_files]

        # Filter resolutions that are smaller than 32
        filtered_resolutions = [res for res in resolutions if res[1] < 32]  # Change res[1] to res[0] for width

        # Separate width and height values
        widths = [res[0] for res in filtered_resolutions]
        heights = [res[1] for res in filtered_resolutions]

        # Add the data to the list
        data.extend([(class_name, width, height) for width, height in zip(widths, heights)])

    # Create a DataFrame from the collected data
    df = pd.DataFrame(data, columns=['Class', 'Width', 'Height'])
    df['Num_Images_Below_32'] = df.groupby('Class')['Width'].transform('count')

    return df

# Example usage:
PATH = "/srv/store/docker-users/pjstuds/obada.kiddeh" # root directory
dataset_directory = PATH + '/FinalDataset/Final_Training/' # path to the dataset
class_resolutions_df = show_image_resolution(dataset_directory)
print(class_resolutions_df)
```

```python
def plot_histogram(df):
    # Group the DataFrame by class and calculate the number of images below 32 pixels
    grouped_df = df.groupby('Class')['Num_Images_Below_32'].count()

    # Plot the histogram
    plt.figure(figsize=(10, 6))
    plt.bar(grouped_df.index, grouped_df.values)
    plt.xlabel('Class')
    plt.ylabel('Number of Images with Resolution Below 32')
    plt.title('Distribution of Image Resolutions Below 32 pixels for Each Class')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
plot_histogram(class_resolutions_df)
```
